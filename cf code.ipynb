### Elbow Method

```python
# Required packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

#### Calculating Elbow method parameters

```python
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(2, 10)
 
for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(df)
    kmeanModel.fit(df)
 
    distortions.append(sum(np.min(cdist(df, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / df.shape[0]) # Distortion List
    inertias.append(kmeanModel.inertia_) #Inertia list
 
    mapping1[k] = sum(np.min(cdist(df, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / df.shape[0]
    mapping2[k] = kmeanModel.inertia_
```

#### The Elbow Method using Inertia

```python
plt.plot(K, inertias, 'bx-') #The Elbow Method Graph using Inertia
#Giving Axis Details
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

From the Elbow Method, we found that Number Of clusters = 4

### Normalisation

```python
from sklearn.preprocessing import StandardScaler
# Scaling the input data to fit the kmeans algorithm
scaler = StandardScaler()
scaler.fit(df)
scaled = scaler.transform(df)
scaled
```

The input values are normalised to apply kmeans clustering Algorithm

### Applying K-Means Clustering Algorithm

```python
from sklearn.cluster import KMeans
#Fitting kmeans algorithm for n_clusters=4
kmeans = KMeans(n_clusters=4)
kmeans.fit(scaled)
```

### Inverse Transformation

```python
# inverse transformation
inversed = scaler.inverse_transform(scaled) # Inverse scaling for the scaled data
print(inversed)
```

After fitting kmeans model, the input datas are rescaled using inverse_transform

### Display the clustering results (not normalised values)

```python
"""Display the clustering results (not normalised values)"""
plt.scatter(inversed[:,0],inversed[:,1], c=kmeans.labels_, cmap='rainbow')
```

When look at the above scatter plot, 4 clusters are formed.
